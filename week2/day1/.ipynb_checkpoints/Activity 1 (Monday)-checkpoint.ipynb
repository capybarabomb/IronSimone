```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import boxcox
pd.options.display.max_rows = 100
## Install xlrd package to load Excel files
#!conda install openpyxl
#!conda install xlrd
```

```python
file1 = pd.read_csv('Data/file1.csv')
file2 = pd.read_csv('Data/file2.csv')
file3 = pd.read_csv('Data/file3.csv')
```

```python
file1.columns
```

```python
file2=file2[['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
file2.columns
```

```python
file3.columns
```

```python
file3.rename(columns={'State':'ST','Gender':'GENDER'}, inplace=True )
file3.columns
```

```python
file3=file3[['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
file3.columns
```

```python
# now that all columns are the same, lets merge the different files

ca_df = pd.concat([file1,file2,file3], axis=0)
ca_df
```

```python
#Standardizing header names

ca_df.rename(columns={'ST':'State','GENDER':'Gender'}, inplace=True )
```

```python
#Removing duplicates and resetting index

ca_df = ca_df.drop_duplicates()
ca_df.reset_index(drop=True, inplace=True)
```

```python
#Deleting and rearranging columns – delete the column customer as it is only a unique identifier for each row of data

ca_df.drop(columns=['Customer'], inplace=True)
```

```python
#Working with data types – Check the data types of all the columns and fix the incorrect ones (for ex. customer lifetime value and number of open complaints )

ca_df.info()
```

```python
# Removing the percentage from the customer lifetime value and truncating it to an integer value.

ca_df['Customer Lifetime Value'] = ca_df['Customer Lifetime Value'].astype('string')
ca_df['Customer Lifetime Value'] = ca_df['Customer Lifetime Value'].str.replace('%','')
ca_df['Customer Lifetime Value'] = ca_df['Customer Lifetime Value'].astype('float')
ca_df['Customer Lifetime Value'] = ca_df['Customer Lifetime Value'].round()
```

```python
ca_df
```

#### Clean the number of open complaints and extract the middle number which is changing between records.


```python
#let's have a look at the serie in question:

print(ca_df['Number of Open Complaints'])
```

```python
# the following function usis regex to extract the number contained between / / in the column 'Number of open complaints'

def extractor():
    
    import re

    for i in range(ca_df.shape[0]):
        if type(ca_df.loc[i, 'Number of Open Complaints']) is str:
            x = re.search(r'/(\d+)/', ca_df.loc[i, 'Number of Open Complaints'])
            if x is not None:
                ca_df.loc[i, 'Number of Open Complaints'] = x.group(1)
    return ca_df['Number of Open Complaints']

extractor()
```

```python
#Filtering data and Correcting typos – Filter the data in state and gender column to standardize the texts in those columns

ca_df['State'].replace('AZ','Arizona', inplace=True)
ca_df['State'].replace('WA','Washington', inplace=True)
ca_df['State'].replace('Cali','California', inplace=True)

ca_df['Gender'].replace('F','Female', inplace=True)
ca_df['Gender'].replace('Femal','Female', inplace=True)
ca_df['Gender'].replace('female','Female', inplace=True)
ca_df['Gender'].replace('M','Male', inplace=True)
```

```python
#exporting the dataframe to a csv file for use in the activity 2.

ca_df.to_csv('ca_df.csv', index=False)
```
